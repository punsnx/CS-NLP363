### 1.2 Custom Tokenizer (add dictionary)

```python
import time
# from pythainlp.tokenize import Tokenizer
from pythainlp.tokenize import word_tokenize
from pythainlp.corpus.common import thai_words
from pythainlp.util import dict_trie

# custom thai dict
cth_dict = set(thai_words())

cth_dict.add("เสริมศักดิ์ พงษ์พานิช")
cth_dict.add("ลาบูบู้")
cth_dict.add("มาสคอต")
cth_dict.add("สุวรรณภูมิ")

trie = dict_trie(dict_source=cth_dict)
Engine = ["attacut","deepcut","longest","newmm","newmm-safe"]
for e in Engine: 
    start = time.time()
    res = word_tokenize(text,engine=e,custom_dict=trie)
    end = time.time()
    print(f"[{e} | Time {end - start:.4f} s] : ",res)
end = time.time()
```

จากการทดลองใช้ algorithms ในการตัดคำ attacut, deepcut, longest,newmm,newmm-safe พบว่าวิธีการตัดคำที่ใช้เวลาน้อยที่สุดคือกลุ่มการตัดคำแบบ dictionary base ซึ่งมีระยะเวลาต่ำกว่า 15 ms ซึ่งการตัดคำแบบ maximal matching มีประสิทธิภาพกว่า longest matching ในกรณีที่ ข้อมูลเยอะมากรุปแบบ longest matching จะใช้เวลานานกว่ามาก และกลุ่ม algorithm deep learning เช่น attacut และ deepcut ใช้เวลาในช่วง 0.5 - 2 s แต่ แบบ deepcut ตัดคำได้ไม่ค่อยตรงมีคำ หรือสระที่เกินแปลกๆ ซึ่ง attacut ค่อนข้างตรงและเร็วกว่า deepcut เทียบได้กับ maximal matching (newmm) ซึ่งจะ match คำให้ได้มากสุด ตัว list tokens ที่ตัดออกมาจะค่อนข้างกระชับและเป็นคำที่มีความหมาย
